# Read dataset

```python
import pandas as pd
from collections import Counter
```

```python
path = 'data/wit_v1.train.all-1percent_sample.tsv'
```

```python
df = pd.read_csv(path, sep='\t')
df
```

## Keep only 5 relevant columns: language, image_url, caption_reference_description, page_title, section_title

```python
df = df[['language', 'image_url', 'caption_reference_description', 'page_title', 'section_title']]
df.drop_duplicates(inplace=True)
df.head()
```

## Drop rows where `caption_reference_description`, `page_title` or `section_title` is NaN

```python
df.dropna(subset=['caption_reference_description', 'page_title', 'section_title'], inplace=True, ignore_index=True)
```

```python
df
```

## Count the number of entries of each language 

```python
languages_count = Counter(df['language'])
languages_count
```

## Only keep the entries in top-20 languages

```python
most_common_langs = [lang for lang, count in languages_count.most_common(20)]
most_common_langs
```

```python
mask = df['language'].isin(most_common_langs)
df_most_common_langs = df[mask]
df_most_common_langs
```

## Equal the number of entries of each language to the least frequent

```python
num_of_least_frequent = languages_count.most_common(20)[-1][1]
num_of_least_frequent
```

```python
sample_dfs = []
for lang in most_common_langs:
    sample_dfs.append(df_most_common_langs[df_most_common_langs['language'] == lang].sample(num_of_least_frequent))
df_sampled = pd.concat(sample_dfs, ignore_index=True)
df_sampled
```

```python
len(df_sampled) == num_of_least_frequent * 20
```

## Save the resulting dataframe as the processed data to further use in **notebook name**

```python
df_sampled.to_csv('data/preprocessed_wit_1_percent_20_langs.tsv', sep='\t')
```
